```python
text = "Text (lateinisch texere ‚weben‘, ‚flechten‘) bezeichnet im nichtwissenschaftlichen Sprachgebrauch eine abgegrenzte, zusammenhängende, meist schriftliche sprachliche Äußerung, im weiteren Sinne auch nicht geschriebene, aber schreibbare Sprachinformation (beispielsweise eines Liedes, Films oder einer improvisierten Theater­aufführung). Aus sprachwissenschaftlicher Sicht sind Texte die sprachliche Form einer kommunikativen Handlung."
```

```python
text
```

```python
list_of_words = text.split()
```

```python
print (list_of_words)
```

```python
list_of_words = [l.lower() for l in list_of_words]

list_of_words = [l.strip('".,;:') for l in list_of_words]
```

```python
list_of_words
```

```python
freqdist ={}
for word in list_of_words:
    if word in freqdist:
        freqdist[word] = freqdist[word]+1
    else:
        freqdist[word] = 1
```

```python
freqdist
```

```python
sorted_freq = dict(sorted(freqdist.items(), key=lambda item: item[1]))
sorted_freq
```
